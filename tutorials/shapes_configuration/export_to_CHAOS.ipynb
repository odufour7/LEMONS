# Export crowd trajectories to ChAOS format
ChAOS (Crowd Animation Open Software) is an animation tool to create crowds of walking characters from a set of 2D trajectories files generated by a crowd-simulation software. Have a look to the [website of the project](https://project.inria.fr/crowdscience/project/ocsr/chaos/).

```python
from configuration.backup import xml_to_Chaos
from pathlib import Path

path_data = Path.cwd().parent.parent / "data"
XML_folder_path = path_data / "xml" / "crowd_ANSURII_tutorial"
filename_CSV = "all_trajectories.csv"
xml_to_Chaos.export_dict_to_CSV(XML_folder_path, filename_CSV)
xml_to_Chaos.export_from_CSV_to_CHAOS(XML_folder_path / filename_CSV, XML_folder_path / "ForChaos", 0.01)
```
