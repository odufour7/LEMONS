# Create a Crowd of pedestrians from the ANSURII dataset
Visit the [database website](https://ph.health.mil/topics/workplacehealth/ergo/Pages/Anthropometric-Database.aspx)

The created crowd is made of agents drawn uniformly from the ANSURII dataset. 

```python
from typing import Final
import matplotlib.pyplot as plt

from shapely.geometry import Polygon

import shapes.utils.constants as constants
from shapes.classes.crowd import Crowd
from shapes.plotting import plot

# Configure plotting (Jupyter magic command)
%matplotlib inline

# Constants
NUMBER_AGENTS: Final[int] = 20
WALL_INTERACTION: Final[bool] = True  # Enable boundary-dependent packing
REPULSION_LENGTH: Final[float] = 15.0  # In centimeters

# Initialize simulation components
boundaries = Polygon([
        (-constants.DEFAULT_BOUNDARY_X, -constants.DEFAULT_BOUNDARY_Y),
        (constants.DEFAULT_BOUNDARY_X, -constants.DEFAULT_BOUNDARY_Y),
        (constants.DEFAULT_BOUNDARY_X, constants.DEFAULT_BOUNDARY_Y),
        (-constants.DEFAULT_BOUNDARY_X, constants.DEFAULT_BOUNDARY_Y)
    ])
crowd = Crowd(boundaries=boundaries)

# Configure crowd agents
crowd.create_agents(number_agents=NUMBER_AGENTS)
crowd.pack_agents_with_forces(
    wall_interaction=WALL_INTERACTION,
    repulsion_length=REPULSION_LENGTH
)

# Visualize results
fig = plot.display_crowd2D(crowd)
plt.show()  # Ensures plot displays in script context
```
